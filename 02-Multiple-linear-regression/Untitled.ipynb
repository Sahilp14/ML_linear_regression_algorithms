```python
import pandas as pd

# Create the dataset
data = {
    'Advertising': [20, 25, 15, 30, 18, 22, 28, 24, 16, 27],
    'StoreSize': [1500, 1800, 1300, 2000, 1400, 1600, 2100, 1700, 1350, 1950],
    'Employees': [10, 12, 8, 14, 9, 11, 13, 12, 9, 13],
    'Sales': [200, 250, 180, 300, 190, 230, 290, 240, 185, 280]
}

df = pd.DataFrame(data)
```

```python
from sklearn.linear_model import LinearRegression

# Independent variables (X) and dependent variable (y)
X = df[['Advertising', 'StoreSize', 'Employees']]
y = df['Sales']

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Coefficients and intercept
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
# Predict sales
y_pred = model.predict(X)

# Compare actual vs predicted
comparison = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
print(comparison)

# Evaluate with R-squared
from sklearn.metrics import r2_score
print("R-squared:", r2_score(y, y_pred))
```

```python
import matplotlib.pyplot as plt

residuals = y - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted Sales")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
