```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('Data.csv')
```

```python
ds.info()
```

```python
x = ds['Age'].mean()
ds['Age'] = ds['Age'].fillna(x)
```

```python
x = ds['Salary'].mean()
ds['Salary'] = ds['Salary'].fillna(x)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
enc_data = ohe.fit_transform(ds[['Country']])

column_names = ohe.get_feature_names_out(['Country'])
enc_df = pd.DataFrame(data = enc_data, columns = column_names)

ds = pd.concat([ds, enc_df], axis=1)
```

```python
ds = ds.drop(columns = ['Country'])
```

```python
ds
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Purchased'] = le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

ds[["Age","Salary"]] = sc.fit_transform(ds[["Age","Salary"]])
```
